```python
class LinearRegression:
    def __init__(self,dim):
        self.coords = []
        self.dim = dim
        self.params = [0]*dim
        self.lr = 3e-4

    def fillCoords(self,coords):
        self.coords = coords

    def getMeanSquaredError(self):
        sum = 0
        for i in range(len(self.coords)):
            output = self.params[-1]
            for j in range(self.dim-1):
                output+=self.params[j]*self.coords[i][j]
            sum += pow(output-self.coords[i][-1],2)
        print(f"The mean squared error is: {sum/(2*len(self.coords))}")
    
    def getDMSE(self):
        d = [0]*self.dim
        for i in range(len(self.coords)):
            output = self.params[-1]
            for j in range(self.dim-1):
                output+=self.params[j]*self.coords[i][j]
            for dim in range(self.dim-1):
                d[dim] += (output-self.coords[i][-1])*self.coords[i][dim]
            d[-1] += (output-self.coords[i][-1])
        for i in d:
            i/=len(self.coords)
        return d

    def train(self,epochs):
        for i in range(epochs):
            derivative = self.getDMSE()
            for j in range(self.dim):
                if derivative[j]>1:
                    derivative[j]=1
                if derivative[j]<-1:
                    derivative[j]=-1
                self.params[j] -= self.lr*derivative[j]
            self.getMeanSquaredError()

    def getLine(self):
        resizedParams = [0]*self.dim
        for i in range(self.dim):
            resizedParams[i] = self.params[i]
        print(resizedParams)

r = LinearRegression(2)
#c = [[30,-10,0],[60,20,5],[120,40,30],[0,10,20]]
c = [[1,0.25],[3.75,0.5],[5,0.75],[8.5,1],[16.5,1.5],[24,2],[30.5,2.5]]
r.fillCoords(c)
r.getMeanSquaredError()
r.train(50000)
r.getLine()
r.getMeanSquaredError()
```
