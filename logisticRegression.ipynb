```python
import math
class logisticRegression:
    def __init__(self,dim):
        self.e = []
        self.w = [0]*(dim-1)
        self.b = 0
        self.lr = 0.01
    
    def fill(self,info):
        self.e = info

    def forwardPass(self,example):
        z=0
        for i in range(len(self.w)):
            z += self.w[i]*self.e[example][i]
        z+=self.b
        out = 1/(1+math.exp(-z))
        return out
    
    def cost(self):
        total = 0
        for ex in range(len(self.e)):
            yHat = self.forwardPass(ex)
            total+=math.pow((yHat-self.e[ex][-1]),2)*0.5
        total/len(self.e)*2
        return total
    
    def derivativeCost(self):
        dw = [0]*len(self.w)
        db = 0
        for ex in range(len(self.e)):
            dz = self.forwardPass(ex)-self.e[ex][-1]
            for i in range(len(self.w)):
                dw[i] += self.e[ex][i]*dz/len(self.e)
            db += dz/len(self.e)
        return dw,db
    
    def train(self,epochs):
        for i in range(epochs):
            print(self.cost())
            dw, db = self.derivativeCost()
            for j in range(len(self.w)):
                self.w[j] -=self.lr*dw[j]
            self.b -=self.lr*db
        print(self.w+[self.b])

    def get(self,x):
        z=0
        for i in range(len(self.w)):
            z += self.w[i]*x[i]
        z+=self.b
        out = 1/(1+math.exp(-z))
        return out
l = logisticRegression(3)
l.fill([[0,0,0],[5,10,0],[10,20,1],[15,30,1]])
l.train(5000)
l.get([7,14])
```
